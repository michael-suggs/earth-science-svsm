```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
dfes = pd.read_csv('RiverRoad_Quan_RF.csv')
dfes
```

```python
labels = ['Decision Tree with Gini', 'Decision Tree with Entropy', 'Bagging', 'Random Forest', 'Ada Boosting with SAMME.R', 'Ada Boosting with SAMME', 'Gradient Boosting', 'Bagging PCA', 'Extreme Gradient Boosting', 'Light Gradient Boosting']
nonwetlands = [.88, .89, .92, .93, .90, .90, .93, .90, .94, .94]
wetlands = [.91, .91, .92, .91, .90, .86, .92, .88, .92, .92]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, nonwetlands, width, label='Non-wetland')
rects2 = ax.bar(x + width/2, wetlands, width, label='Wetland')

ax.set_ylabel('Precision')
ax.set_title('Comparing Models Precision Values for Wetland vs Non-Wetland')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

ax.set_ylim([.70, 1])
plt.setp(ax.xaxis.get_majorticklabels(), rotation=270)
plt.grid(True, alpha=.5)

plt.show()
```
